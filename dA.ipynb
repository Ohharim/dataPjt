```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm 
import numpy as np 
import numpy as np
import pandas as pd

print(mpl.matplotlib_fname())
print(mpl.get_cachedir())
mpl.rcParams['font.family'] = 'NanumSquare Neo variable'
```

```python

df1 = pd.read_csv('__.csv') #
df2 = pd.read_csv('__.csv')

plt.figure(figsize=(12, 6))

x1 = np.arange(len(df1))#
plt.barh(x1 - 0.15, df1['  '], label='  ', color='blue', height=0.2)
plt.barh(x1 - 0.15, df1['  '], left=df1['  '], label='  ', color='green', height=0.2)
plt.barh(x1 - 0.15, df1[' '], left=df1['  '] + df1['  '], label=' ', color='yellow', height=0.2)
plt.barh(x1 - 0.15, df1[' '], left=df1['  '] + df1['  '] + df1[' '], label=' ', color='red', height=0.2)

x2 = np.arange(len(df2))#
plt.barh(x2 + 0.15, df2['  '], label='  ', color='blue', height=0.2,alpha=0.3)
plt.barh(x2 + 0.15, df2['  '], left=df2['  '], label='  ', color='green', height=0.2,alpha=0.3)
plt.barh(x2 + 0.15, df2[' '], left=df2['  '] + df2['  '], label=' ', color='yellow', height=0.2,alpha=0.3)
plt.barh(x2 + 0.15, df2[' '], left=df2['  '] + df2['  '] + df2[' '], label=' ', color='red', height=0.2,alpha=0.3)

x = np.arange(max(len(df1), len(df2)))

year = [2018, 2019, 2020, 2021]
plt.yticks(x, year)
plt.xticks([0,2,15,18,20,40,60,80,84,100])
plt.ylabel('')
plt.xlabel('%')
plt.title('    (2018-2021)')
plt.grid(axis='x')
plt.legend(['  ','  ',' ',' '],loc='center left', bbox_to_anchor=(1, 0.5))
for i, j in enumerate(x1) :
    plt.text(i+1, j-0.35, '', ha = 'left')
for i, j in enumerate(x1) :
    plt.text(i+1, j+0.3, '', ha = 'left')
plt.show()
```

```python


df = pd.read_csv('ADHD.csv', index_col=0)

plt.figure(figsize=(10, 6))  

for city in df.index:
    plt.plot(df.columns, df.loc[city], label=city, marker='o')

plt.title(' ADHD (2018-2022)')
plt.xlabel('')
plt.ylabel('')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='x')
plt.show()
```

```python


df = pd.read_csv('ADHD.csv', index_col=0)

excluded_indexes = [0, 1, 7]

plt.figure(figsize=(10, 6)) 

for city in df.index:
    if df.index.get_loc(city) not in excluded_indexes:
        plt.plot(df.columns, df.loc[city], label=city, marker='o')

plt.title(', ,    ADHD (2018-2022)')
plt.xlabel('')
plt.ylabel('')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='x')
plt.show()
```

```python
data = pd.read_csv('_2020.csv')

plt.figure(figsize=(10, 6))

for index, row in data.iterrows():
    plt.bar(row['Unnamed: 0'], row[''], label='', color='red', alpha=0.7)
    plt.bar(row['Unnamed: 0'], row[' '], bottom=row[''], label=' ', color='yellow', alpha=0.7)
    plt.bar(row['Unnamed: 0'], row[''], bottom=row['']+row[' '], label='', color='blue', alpha=0.7)

plt.xlabel('')
plt.ylabel('%')
plt.title('   (2020)')
plt.legend(['',' ',''],loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

plt.show()

```
