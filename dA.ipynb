```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm 
import numpy as np 
import numpy as np
import pandas as pd

print(mpl.matplotlib_fname())
print(mpl.get_cachedir())
mpl.rcParams['font.family'] = 'NanumSquare Neo variable'
```

```python

df1 = pd.read_csv('과의존_심각성_과의존위험군.csv') #인식
df2 = pd.read_csv('과의존_심각성_일반.csv')

plt.figure(figsize=(12, 6))

x1 = np.arange(len(df1))#위험
plt.barh(x1 - 0.15, df1['전혀 심각하지 않다'], label='전혀 심각하지 않다', color='blue', height=0.2)
plt.barh(x1 - 0.15, df1['별로 심각하지 않다'], left=df1['전혀 심각하지 않다'], label='별로 심각하지 않다', color='green', height=0.2)
plt.barh(x1 - 0.15, df1['다소 심각하다'], left=df1['전혀 심각하지 않다'] + df1['별로 심각하지 않다'], label='다소 심각하다', color='yellow', height=0.2)
plt.barh(x1 - 0.15, df1['매우 심각하다'], left=df1['전혀 심각하지 않다'] + df1['별로 심각하지 않다'] + df1['다소 심각하다'], label='매우 심각하다', color='red', height=0.2)

x2 = np.arange(len(df2))#일반
plt.barh(x2 + 0.15, df2['전혀 심각하지 않다'], label='전혀 심각하지 않다', color='blue', height=0.2,alpha=0.3)
plt.barh(x2 + 0.15, df2['별로 심각하지 않다'], left=df2['전혀 심각하지 않다'], label='별로 심각하지 않다', color='green', height=0.2,alpha=0.3)
plt.barh(x2 + 0.15, df2['다소 심각하다'], left=df2['전혀 심각하지 않다'] + df2['별로 심각하지 않다'], label='다소 심각하다', color='yellow', height=0.2,alpha=0.3)
plt.barh(x2 + 0.15, df2['매우 심각하다'], left=df2['전혀 심각하지 않다'] + df2['별로 심각하지 않다'] + df2['다소 심각하다'], label='매우 심각하다', color='red', height=0.2,alpha=0.3)

x = np.arange(max(len(df1), len(df2)))

year = [2018, 2019, 2020, 2021]
plt.yticks(x, year)
plt.xticks([0,2,15,18,20,40,60,80,84,100])
plt.ylabel('연도')
plt.xlabel('%')
plt.title('스마트폰 과의존 심각성에 대한 인식(2018-2021)')
plt.grid(axis='x')
plt.legend(['전혀 심각하지 않다','별로 심각하지 않다','다소 심각하다','매우 심각하다'],loc='center left', bbox_to_anchor=(1, 0.5))
for i, j in enumerate(x1) :
    plt.text(i+1, j-0.35, '과의존위험군', ha = 'left')
for i, j in enumerate(x1) :
    plt.text(i+1, j+0.3, '일반사용자군', ha = 'left')
plt.show()
```

```python
df1 = pd.read_csv('과의존_심각성_과의존위험군.csv') #전혀심각하지않음 강조
df2 = pd.read_csv('과의존_심각성_일반.csv')

plt.figure(figsize=(10, 6))
x = np.arange(len(df1))

plt.plot(x, df1['전혀 심각하지 않다'], label='과의존위험군', color='red', marker='o')
plt.plot(x, df2['전혀 심각하지 않다'], label='일반사용자군', color='blue', marker='s')

year = [2018, 2019, 2020, 2021]
plt.xticks(x, year)
plt.xlabel('연도')
plt.ylabel('%')

plt.title('스마트폰 과의존 심각성 인식 중 \'전혀 심각하지 않다\'')
plt.grid(axis='x')

plt.legend()
plt.show()
```

```python


df = pd.read_csv('지역별ADHD.csv', index_col=0)

plt.figure(figsize=(10, 6))  

for city in df.index:
    plt.plot(df.columns, df.loc[city], label=city, marker='o')

plt.title('시도별 ADHD 진료현황(2018-2022)')
plt.xlabel('연도')
plt.ylabel('명')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='x')
plt.show()
```

```python


df = pd.read_csv('지역별ADHD.csv', index_col=0)

excluded_indexes = [0, 1, 7]

plt.figure(figsize=(10, 6)) 

for city in df.index:
    if df.index.get_loc(city) not in excluded_indexes:
        plt.plot(df.columns, df.loc[city], label=city, marker='o')

plt.title('서울, 경기, 부산 제외 시도별 ADHD 진료현황(2018-2022)')
plt.xlabel('연도')
plt.ylabel('명')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='x')
plt.show()
```

```python
data = pd.read_csv('코로나이후_일상변화2020.csv')

plt.figure(figsize=(10, 6))

for index, row in data.iterrows():
    plt.bar(row['Unnamed: 0'], row['증가'], label='증가', color='red', alpha=0.7)
    plt.bar(row['Unnamed: 0'], row['변화 없음'], bottom=row['증가'], label='변화 없음', color='yellow', alpha=0.7)
    plt.bar(row['Unnamed: 0'], row['감소'], bottom=row['증가']+row['변화 없음'], label='감소', color='blue', alpha=0.7)

plt.xlabel('항목')
plt.ylabel('%')
plt.title('코로나 이후 일상 변화(2020)')
plt.legend(['증가','변화 없음','감소'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

plt.show()

```
